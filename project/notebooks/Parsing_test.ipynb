```python
import unicodedata

def remove_accents(input_str):
    nkfd_form = unicodedata.normalize('NFKD', unicode(input_str))
    return u"".join([c for c in nkfd_form if not unicodedata.combining(c)])
```

```python
def partition_title(m):
    m=remove_accents(m)
    #print(m)
    m=m.encode('utf8')
    t1=m.partition('.')
    rank=t1[0]
    t2=t1[2].partition('–')
    artist=t2[2]
    title=t2[0].strip()
    title=re.sub('"','',title)
    title=re.sub('”','',title)
    rank=rank.strip()
    artist=artist.strip()
    title =title.strip()
    print(rank,title,artist)
    return rank, artist,title
    
```
