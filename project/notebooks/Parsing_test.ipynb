```python
def partition_title(m):
    m=m.encode('utf-8')
    t1=m.partition('.')
    rank=t1[0]
    t2=t1[2].partition('–')
    artist=t2[2]
    title=re.sub('"','',t2[0])
    title=re.sub('”','',title)
    rank=rank.strip()
    artist=artist.strip()
    title =title.strip()
    print(rank, artist,title)    
```
